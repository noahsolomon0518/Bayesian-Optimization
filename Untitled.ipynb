```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor as GPR
from sklearn.gaussian_process.kernels import RBF

#Create perfect line of data
def createData(w,b):
    x = [i for i in range(20)]
    y = [w*xi+b for xi in x]
    plt.scatter(x,y)
    return (x,y)
x,y = createData(2,5)
```

```python
def negativeMSE(w, b, xTrain, yTrain):
    xTrain = np.array(xTrain)
    yTrain = np.array(yTrain)
    yPred = w*xTrain+b
    SSE = sum((np.array(yPred)-np.array(yTrain))**2)
    NMSE = -SSE/len(yPred)
    return NMSE



#The weights that we are aiming for. Results in a -MSE = 0 because the data is a perfect line.
print(negativeMSE(2,5,x,y))
```

```python
def MEI(predNMSE,std, domain):
    ind = np.argmax(np.array(predNMSE)+3*np.array(std))
    return np.array(domain)[ind]



#model 3 has the highest NMSE but the high uncertainty with model 2 leads to a larger expected improvement
print(MEI([-2000,-10,-5], [200,3,1], [['model1 parameters'],['model2 parameters'], ['model3 parameters']]))
```

```python
def calculateWeights(x,y, iterations):
    domain = [(w,b) for w in range(20) for b in range(20)]
    params = []
    accuracies = []
    curParams = (0,0)
    for i in range(iterations):
        NMSE = negativeMSE(curParams[0],curParams[1], x,y)
        params.append([curParams[0],curParams[1]])
        accuracies.append(NMSE)
        gprModel = GPR(kernel=RBF(),random_state=0).fit(params, accuracies)
        predNMSE, std = gprModel.predict(domain,return_std = True)
        print("weights: ", curParams, "NMSE: ",NMSE )
        curParams = MEI(predNMSE, std, domain)
    print("Final weights are: ", curParams)
    return (params,accuracies)
```

```python
params,accuracies = calculateWeights(x,y,100)
```

```python
from mpl_toolkits.mplot3d import Axes3D
params = np.array(params)
accuracies = np.array(accuracies)


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(params[:,0],params[:,1],accuracies)
ax.scatter(2,5,0, c = 'yellow', s = 5)
ax.set_xlabel('w', c = 'b')
ax.set_ylabel('b', c = 'b')
ax.set_zlabel('NMSE', c = 'b')
ax.set_title('NMSE with respect to hyperparameters')
```
